# Question is 
IS THERE ANY RELATIONSHIP BETWEEN CORONA VIRUS IN OUR COUNTRY AND HOW HAPPY PEOPLE LIVING IN PERTICULAR COUNTRY

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
dataset = pd.read_csv('Covid19_Confirmed_dataset.csv')
dataset.head()
```

```python
dataset.shape
```

```python
useless_col = ['Province/State' , 'Lat' , 'Long']
```

```python
dataset.drop(useless_col , axis = 1, inplace = True)
dataset
```

```python
corona_dataset = dataset.groupby('Country/Region').sum()
```

```python
corona_dataset
```

```python
corona_dataset.shape
```
